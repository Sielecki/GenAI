```python
import ollama 
```

```python
class Agent:
    def __init__(self, name, personality, model='llama3.2', rag_engine=None):
        self.name = name
        self.personality = personality
        self.model = model
        self.memory = []
        self.rag_engine = rag_engine
        self.system_prompt = (
            f"You are {name}, a character with the following traits: {personality}\n"
            "I want to act according to this character to the world events and the dialogues from other companions. Please use dialogue and add a possible action.\n"
            "You are a character in a story. You can only see the world events and the dialogues from other companions.\n"
            "Please avoid using the same responses and continue the story as you see fit\n"
            )
        ollama.chat(
            model=self.model,
            messages=[
                {'role': 'system', 'content': self.system_prompt}
            ]
        )

    def prompt_agent(self, user="", user_input="", world_state=""):

        self.memory.append(f"World Event: {world_state}")
        self.memory.append(f"{user}: {user_input}")
        
        prompt = "Conversation so far:\n"
        for m in self.memory:
            prompt += f"{m}\n"
        prompt += f"Current world event: {world_state}\n"
        prompt += f"{user}: {user_input}\n{self.name}:"

        response = ollama.chat(
            model=self.model,
            messages=[
                {'role': 'user', 'content': prompt}
            ]
        )
        reply = response['message']['content']
        
        self.memory.append(f"{self.name}: {reply}")
        return reply, self.name
    
    def get_memory(self):
        return self.memory
```

```python
hero = Agent(
    name="Aegon the Brave",
    personality="Courageous, curious, slightly overconfident warrior from the Northlands."
)

dog = Agent(
    name="Zoro the Talking Dog",
    personality="Slightly cowardly, loyal, skilled in battle, stutters."
)
world_event = "A dragon has just landed on the outskirts of the village."
response, character_name = hero.prompt_agent("Narator", "What should we do?", world_event)
print(f"Hero: {response}")
response, character_name = dog.prompt_agent(character_name, response, world_event)
print(f"Dog: {response}")

world_event = "The dragon started to attack the village."
print(f"World Event: {world_event}")
response, character_name = hero.prompt_agent(character_name, response, world_event)
print(f"Hero: {response}")
response, character_name = dog.prompt_agent(character_name, response, world_event)
print(f"Dog: {response}")

world_event = "The dragon is on a rampage and starts to kill villagers."
print(f"World Event: {world_event}")
response, character_name = dog.prompt_agent(character_name, response, world_event)
print(f"Dog: {response}")
response, character_name = hero.prompt_agent(character_name, response, world_event)
print(f"Hero: {response}")

world_event = "Suddnely, the dragon stops and offers you a deal. 'Dragom: I will spare the village if you bring me the golden crown from the castle.'"
print(f"World Event: {world_event}")
response, character_name = hero.prompt_agent(character_name, response, world_event)
print(f"Hero: {response}")
```
